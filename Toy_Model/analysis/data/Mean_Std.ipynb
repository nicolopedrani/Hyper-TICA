## M&uuml;ller Potential 
Just have a look at the Muller potential

```python
def plot_result(filename="data/mean1.txt",title='Deep-TICA 1'):
    #-- prepare grid points
    y = np.linspace(-0.3,2.1,150)
    x = np.linspace(-1.8,1.2,150)
    X, Y = np.meshgrid(x, y)
    Z = np.loadtxt(filename)

    #-- ax for plotting 
    fig,ax = plt.subplots(figsize=(8,8)) 

    #-- plot isolines 
    cset = ax.contourf(X,Y,Z,linewidths=1,cmap="fessa")
    fig.colorbar(cset, ax=ax)
    cset = ax.contour(X,Y,Z,linewidths=3,cmap="gray",linestyles="dashed")
    ax.clabel(cset,inline=True,fmt='%1.1f',fontsize=15)
    ax.set_xlabel("p.x")
    ax.set_ylabel("p.y")
    ax.set_title(title)

    plt.tight_layout()
```

```python
plot_result("data/mean1.txt","Deep TICA 1")
```

```python
plot_result("data/mean2.txt","Deep TICA 2")
```

```python
plot_result("data/std1.txt","DevStd Deep TICA 1")
```

```python
plot_result("data/std2.txt","DevStd Deep TICA 2")
```

```python
plot_result("data/var1.txt","Variance Deep TICA 1")
```

```python
plot_result("data/var2.txt","Variance Deep TICA 2")
```
